### Задание: 
- построить систему распознавания геометрических фигур на изображении, которая:
    - получая на вход картинку, выделяет на ней все геометрические фигуры (круг, треугольник, квадрат, прямоугольник) и подписывает их;
    - позволяет оценить правильность определения фигур и точность классификации.

### Исходные данные:
- Набор из 15 изображений (Samples/), на которых изображены геометрические фигуры и присутсвует шум и неравномерное фоновое освещение.


### Возможные блоки, из которых может состоять пайплайн:
- блок считывания изображения;
- блок предварительной обработки изображения;
- блок сегментации фигур;
- блок выделения признаков;
- блок подготовки обучающей выборки;
- блок классификатора;
- блок оценки точности пайплайна;
- блок визуализации результатов.
