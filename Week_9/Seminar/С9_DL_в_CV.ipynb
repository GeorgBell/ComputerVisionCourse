## 1.3. В чем заключается обучение?

- **Обучение с учителем** — это тип тренировок присущий таким проблемам как регрессия и классификация. В этом случае, обучающая выборка содержит аннотированные данные, т.е. каждому вектору признаков соответствует метка класса, являющаяся желаемым результатом классификации. Нейронной сети предоставляются входные данные и желаемый результат, то есть ученик-ИНС, посмотрев на входные данные поймет, что нужно стремиться к тому результату который ему предоставили.
- **Обучение без учителя** — этот тип обучения встречается реже. В этом слу-чае нет учителя-меток, поэтому сеть не получает желаемый результат. В основ-ном такой вид тренировок присущ НС у которых задача состоит в группировке данных по определенным параметрам (задача кластеризации). 
- Если же у исследователя есть небольшая обучающая выборка с аннотированными данными, а большая часть данных не аннотирована, то применяется подход обучения с **частичным привлечением учителя**.
- Существует метод **обучения с подкреплением** (reinforcement learning). Такой способ применим, когда, основываясь на результатах полученных от ИНС, можно дать ей оценку: вознаграждение или штраф. Например, существует ИНС, обученная играть в видеоигры (PAC-MAN). Тогда каждый раз, когда НС будет набирать много очков в игре, она будет получать поощрение. Иными словами, нейронной сети предоставляется право найти любой способ достижения цели до тех пор, пока он будет давать хороший результат.   
___
В классическом случае, обучение нейронной сети заключается в:
- распространении сигнала по графу и вычислении сигналов на выходе каждого нейрона;
- вычисление ошибки для каждого нейрона путем метода обратного распространения;
- изменение весов связей.

## 1.4. Применение, ограничения, перспективы
**Применения:**
- анализ изображений и видео (сегментация, классификация, детекция);
- обработка естественного языка / Natural Language Processing (языковые модели, фонетические модели);

- синтез речи;

- требование к большим объемам размеченных данных (для обучения с учителем);

- ограничения в переносе обучения (transfer learning);
- отсутствие четких стратегий и алгоритмов разработки DL-решений.
___
**Перспективы:**
- большой простор для комбинации различных подходов и архитектур;
- смещение фокуса AI-сообщества на модели с обучением без учителя;
- алгоритмы для работы с малыми данными (Small Data);
- интерпретируемое машинное обучение и объяснимый искусственный интеллект;
- совершенствование генеративных моделей.

## 2.1. Перцептрон
### 2.1.1 Теоретическое описание  
Перцептрон - простейший тип искусственной нейронной сети, представляющий одиночный нейрон. Это модель одиночного нейрона, которую можно использовать для задач классификации двух классов, она обеспечивает основу для дальнейшего развития гораздо более крупных сетей.  
Приницп работы перцептрона вдохновлен обработкой информации одной нервной клеткой. Нейрон принимает входные сигналы через свои дендриты, которые передают электрический сигнал телу клетки. Точно так же перцептрон получает входные сигналы из примеров обучающих данных, которые мы взвешиваем и объединяем в линейное уравнение:
- x - входные данные. Связь с нейроном имеет вес w. Каждый раз, когда значение проходит через соединение, вы умножаете это значение на вес соединения. Для входа x нейрон достигает значение w * x. Нейронная сеть «учится», изменяя свои веса;
- B - это особый вид веса, который мы называем смещением. Смещение не имеет связанных с ним входных данных; смещение позволяет нейрону изменять выходной сигнал независимо от его входов;
- Y - это значение, которое в конечном итоге выводит нейрон. Чтобы получить результат, нейрон суммирует все значения, которые он получает через свои соединения. Активация этого нейрона y = w * x + b.

Таким образом, перцептрон представляет собой алгоритм классификации для задач с двумя классами (0 и 1), где линейное уравнение (гиперплоскость) может использоваться для разделения двух классов.
Это тесно связано с линейной регрессией и логистической регрессией, которые делают прогнозы аналогичным образом (взвешиванием сумма входных данных).
___
Важным этапом является обучение перцептрона, которое заключается в решении задачи оптимизации весов на основе обучающих данных с использованием стохастического градиентного спуска.  
**Градиентный спуск** - это процесс минимизации функции потерь, который включает в себя расчет целевой функции (функции потерь), ее производной для определения направления поиска оптимального значения.  
Простешим вариантом градиентного алгоритма оптимизации являеся **стохастический градиентный спуск**. Принцип работы этого алгоритма оптимизации заключается в том, что каждый обучающий экземпляр отображается модели по одному. Модель делает прогноз для обучающего экземпляра, вычисляется ошибка и веса модели обновляются, чтобы уменьшить ошибку для следующего прогнозирования.

В качестве учебного примера используем простой [бинарный датасет по раку молочной железы](https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic)) на основании признаков, полученных из оцифрованных изображений тонкоигольного аспирата (FNA) новообразования молочной железы;

```python
# Используем библиотеку машинного обучения для загрузки стандартных датасетов
import sklearn.datasets
from sklearn.model_selection import train_test_split # для разделения выборки
from sklearn.metrics import (confusion_matrix, # матрица несоответствия
                             classification_report, # сводка по классификации
                             accuracy_score) # оценка точности
from sklearn.preprocessing import MinMaxScaler # Нормирование входных данных

import numpy as np # Библиотека векторных вычислений
import matplotlib.pyplot as plt # Библиотека построения графиков
import pandas as pd # Библиотека для статистической обработки табличных данных
plt.style.use("seaborn")
```

- Посмотрим на данные;

```python
# Загружаем учебный набор данных
breast_cancer = sklearn.datasets.load_breast_cancer()

# Преобразуем данные в табличный формат pandas
data = pd.DataFrame(breast_cancer.data, columns = breast_cancer.feature_names)
data["class"] = breast_cancer.target
data.head()
```

```python
# Основные статистические параметры
data.describe()
```

```python
# Построим график распределения данных по классам
data['class'].value_counts().plot(kind = "barh")
plt.xlabel("Count")
plt.ylabel("Classes")
plt.show()
```

- Подготовим обучающий и тестовый датасет

```python
# Подготовим входные данные (вектор признаков)
X = data.drop("class", axis = 1)

# Проведем нормирование входных данных
mnscaler = MinMaxScaler()
X = mnscaler.fit_transform(X)

# Подготовим истинные значения выходных откликов
Y = data["class"]

# Разделим набор данных на обучающую и тестовую выборки
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state = 1)
```

- Реализуем класс Перцептрон и его методы;

```python
class Perceptron:
    def __init__ (self):
        # Исходная функция, инициализирует веса
        self.w = None
        self.b = None
    
    def model(self, x):
        # Модель перцептрона, реализующая логику Y' = W*X + b, 
        # Y_pred = 1 if Y' > 0, else 0
        return 1 if (np.dot(self.w, x) - self.b >= 0) else 0
    
    def predict(self, X):
        # Применение модели для получения отклика
        Y = []
        for x in X:
            result = self.model(x)
            Y.append(result)
        return np.array(Y)
    
    def fit(self, X, Y, epochs = 1, lr = 1):
        # Обучение модели

        # Присваивание стартовых значений весам
        self.w = np.ones(X.shape[1])
        self.b = 0

        accuracy = {}
        max_accuracy = 0

        wt_matrix = []

        # Итерируемся по шагам обучения
        for i in range(epochs):
            for x, y in zip(X, Y):
                # Получаем отклик модели
                y_pred = self.model(x)
                # Обновление весов в зависимости от результата
                if y == 1 and y_pred == 0:
                    self.w = self.w + lr * x
                    self.b = self.b - lr * 1
                elif y == 0 and y_pred == 1:
                    self.w = self.w - lr * x
                    self.b = self.b + lr * 1

            # Сохранение нового набора весов
            wt_matrix.append(self.w)    
            accuracy[i] = accuracy_score(self.predict(X), Y)
            if (accuracy[i] > max_accuracy):
                max_accuracy = accuracy[i]
                j = i
                chkptw = self.w
                chkptb = self.b

        self.w = chkptw
        self.b = chkptb

        # Построение графика обучения
        print(max_accuracy,j)
        plt.plot(list(accuracy.values()))
        plt.xlabel("Epoch #")
        plt.ylabel("Accuracy")
        plt.ylim([0, 1])
        plt.show()
```

- Обучим модель;

```python
perceptron_model = Perceptron()
perceptron_model.fit(X_train, Y_train, epochs = 100)
```

- Протестируем модель;

```python
Y_pred = perceptron_model.predict(X_test)
Y_pred
```

```python
print(classification_report(Y_test, Y_pred))
print(confusion_matrix(Y_test, Y_pred))
```

### 2.1.3. Используем библиотеку для создания перцептрона
Для эффективности и повышения скорости разработки нейронных сетей применяются специализированные и оптимизированные программные библиотеки:

- Импортируем библиотеки;

```python
from tensorflow import keras
from tensorflow.keras import layers
```

- Создадим перцептрон;

```python
X_train.shape
```

```python
# Create a network with 1 linear unit
perceptron_model_keras = keras.Sequential([
    layers.Dense(units=1, input_shape=[30])
])
```

```python
perceptron_model_keras.compile(
    optimizer='adam', # Используем мрдифицированный алгоритм оптимизации ADAM
    loss='binary_crossentropy', # Используем бинарную кросс-энтропию для расчета ошибки
    metrics=['acc']) # В качестве дополнительной метрики используем точность
```

## 2.2. Полносвязная нейронная сеть
### 2.2.1 Теоретическое описание  

Ключевой идеей нейронной сети является модульность: построение сложной сети из более простых функциональных единиц. Мы видели, как нейрон вычисляет линейную функцию - теперь мы увидим, как комбинировать и изменять эти отдельные блоки для моделирования более сложных отношений.  
Нейроны образуют слои нейронной сети. Когда мы собираем вместе линейные блоки, имеющие общий набор входных данных, мы получаем полносвязный слой.

- Зададим архитектуру нейроонной сети

```python
nn_architecture = [
    {"layer_size": 3, "activation": "none"}, # input layer
    {"layer_size": 5, "activation": "relu"},
    {"layer_size": 4, "activation": "relu"},
    {"layer_size": 3, "activation": "relu"},
    {"layer_size": 1, "activation": "sigmoid"}
]
```

Инициализация параметров с использованием небольших случайных чисел - простой подход, который гарантирует достаточно хорошую отправную точку для алгоритма оптимизации. 
- Напишем функцию инициализации начальных значений весов

```python
def initialize_parameters(nn_architecture, seed = 3):
    np.random.seed(seed)
    # python dictionary containing our parameters "W1", "b1", ..., "WL", "bL"
    parameters = {}
    number_of_layers = len(nn_architecture)

    for l in range(1, number_of_layers):
        parameters['W' + str(l)] = np.random.randn(
            nn_architecture[l]["layer_size"],
            nn_architecture[l-1]["layer_size"]
            ) * 0.01
        parameters['b' + str(l)] = np.zeros((nn_architecture[l]["layer_size"], 1))
        
    return parameters
```

Функции активации придают нейронным сетям нелинейность. В нашем примере мы будем использовать сигмоид и ReLU.
Сигмоид выводит значение от 0 до 1, что делает его очень хорошим выбором для бинарной классификации.
- Напишем функции для для расчета функции активации и их производной, которая пригодится для обратного распространения ошибки

```python
def sigmoid(Z):
    S = 1 / (1 + np.exp(-Z))
    return S

def relu(Z):
    R = np.maximum(0, Z)
    return R

def sigmoid_backward(dA, Z):
    S = sigmoid(Z)
    dS = S * (1 - S)
    return dA * dS

def relu_backward(dA, Z):
    dZ = np.array(dA, copy = True)
    dZ[Z <= 0] = 0
    return dZ
```

- Напишем функции для расчета прямого распространения

```python
# Расчет произведения матриц
def linear_forward(A, W, b):
    Z = np.dot(W, A) + b
    return Z

# Применение функции активации
def linear_activation_forward(A_prev, W, b, activation):
    if activation == "sigmoid":
        Z = linear_forward(A_prev, W, b)
        A = sigmoid(Z)
    elif activation == "relu":
        Z = linear_forward(A_prev, W, b)
        A = relu(Z)
    return Z, A

# Построение L-слойной модели
def L_model_forward(X, parameters, nn_architecture):
    forward_cache = {}
    A = X
    number_of_layers = len(nn_architecture)
    
    for l in range(1, number_of_layers):
        A_prev = A 
        W = parameters['W' + str(l)]
        b = parameters['b' + str(l)]
        activation = nn_architecture[l]["activation"]
        Z, A = linear_activation_forward(A_prev, W, b, activation)
        forward_cache['Z' + str(l)] = Z
        forward_cache['A' + str(l-1)] = A_prev
    AL = A
    return AL, forward_cache
```

Мы используем «кеш» для передачи переменных, вычисленных во время прямого распространения, на соответствующем шаге обратного распространения. Данный кеш содержит значения, используемые для обратного распространения при вычислении производных.

Чтобы отслеживать процесс обучения, определять расхождение между предсказаниями моделей и правильными ответами, необходимо вычислить значение функции стоимости. Для задачи бинарной классификации применяют функцию бинарной кросс-энтропии.

- Напишем функцию для расчета кросс-энтропии

```python
def compute_cost(AL, Y):
    m = Y.shape[1]

    # Compute loss from AL and y
    logprobs = np.multiply(np.log(AL),Y) + np.multiply(1 - Y, np.log(1 - AL))
    # cross-entropy cost
    cost = - np.sum(logprobs) / m

    cost = np.squeeze(cost)
    
    return cost
```

Обратное распространение используется для вычисления градиента функции потерь относительно параметров. Этот алгоритм оптимизации представляет собой рекурсивное использование «цепного правила» дифференцирования.

- Напишем функции для расчета обратного распространения

```python
def L_model_backward(AL, Y, parameters, forward_cache, nn_architecture):
    grads = {}
    number_of_layers = len(nn_architecture)
    m = AL.shape[1]
    Y = Y.reshape(AL.shape) # after this line, Y is the same shape as AL
    
    # Initializing the backpropagation
    dAL = - (np.divide(Y, AL) - np.divide(1 - Y, 1 - AL))
    dA_prev = dAL

    for l in reversed(range(1, number_of_layers)):
        dA_curr = dA_prev

        activation = nn_architecture[l]["activation"]
        W_curr = parameters['W' + str(l)]
        Z_curr = forward_cache['Z' + str(l)]
        A_prev = forward_cache['A' + str(l-1)]

        dA_prev, dW_curr, db_curr = linear_activation_backward(dA_curr, Z_curr, A_prev, W_curr, activation)

        grads["dW" + str(l)] = dW_curr
        grads["db" + str(l)] = db_curr

    return grads

def linear_activation_backward(dA, Z, A_prev, W, activation):
    if activation == "relu":
        dZ = relu_backward(dA, Z)
        dA_prev, dW, db = linear_backward(dZ, A_prev, W)
    elif activation == "sigmoid":
        dZ = sigmoid_backward(dA, Z)
        dA_prev, dW, db = linear_backward(dZ, A_prev, W)

    return dA_prev, dW, db

def linear_backward(dZ, A_prev, W):
    m = A_prev.shape[1]

    dW = np.dot(dZ, A_prev.T) / m
    db = np.sum(dZ, axis=1, keepdims=True) / m
    dA_prev = np.dot(W.T, dZ)

    return dA_prev, dW, db
```

- Напишем функцию для обновления параметров

```python
def update_parameters(parameters, grads, learning_rate):
    L = len(parameters) // 2 # number of layers in the neural network

    for l in range(1, L):
        parameters["W" + str(l)] = parameters["W" + str(l)] - learning_rate * grads["dW" + str(l)]
        parameters["b" + str(l)] = parameters["b" + str(l)] - learning_rate * grads["db" + str(l)]

    return parameters
```

- Напишем функцию, создающую модель, на основе написанных ранее методов

```python
def L_layer_model(X, Y, nn_architecture, learning_rate = 0.0075, num_iterations = 3000, print_cost=False):
    np.random.seed(1)
    # keep track of cost
    costs = []
    
    # Parameters initialization.
    parameters = initialize_parameters(nn_architecture)
    
    # Loop (gradient descent)
    for i in range(0, num_iterations):

        # Forward propagation: [LINEAR -> RELU]*(L-1) -> LINEAR -> SIGMOID.
        AL, forward_cache = L_model_forward(X, parameters, nn_architecture)
        
        # Compute cost.
        cost = compute_cost(AL, Y)
    
        # Backward propagation.
        grads = L_model_backward(AL, Y, parameters, forward_cache, nn_architecture)
 
        # Update parameters.
        parameters = update_parameters(parameters, grads, learning_rate)
                
        # Print the cost every 100 training example
        if print_cost and i % 100 == 0:
            print("Cost after iteration %i: %f" %(i, cost))

        costs.append(cost)
            
    # plot the cost
    plt.plot(np.squeeze(costs))
    plt.ylabel('cost')
    plt.xlabel('iterations (per tens)')
    plt.title("Learning rate =" + str(learning_rate))
    plt.show()
    
    return parameters
```

## 2.3. Сверточная нейронная сеть

Сверточная нейронная сеть, используемая для задачи классификации изображений, состоит из двух частей: сверточной основы и полносвязной надстройки.

Сверточные слои используются для извлечения признаков из изображения. Полносвязная надстройка используется для определения класса изображения на основе выделенных признаков. 

В качестве выделяемых признаков в порядке иерархического усложнения может служить линия, текстура, форма, узор, более сложные геометрические объекты.

Следующим важным структурным элементом является слой пулинга или субдискретизации. Этот слой позволяет снижать размерность данных.

Вместо этого, применяя пулинг, мы сжимаем карту признаков, чтобы выделить только самую полезную часть.

Наиболее распространенным подходом является max pooling: он заключается в выборе максимального значения в каждой карте признаков.

### 2.3.2 Реализация СНС на основе фреймворка  

- Проверим формат входных данных

```python
data_cnc = data_merged
data_cnc.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data_cnc, labels_merged, 
                                                    test_size = 0.2,
                                                    random_state = 42)
```

```python
cnc_model = keras.Sequential([

    # First Convolutional Block
    layers.Conv2D(filters=32, kernel_size=5, activation="relu", padding='same',
                  input_shape=[X_train.shape[1], X_train.shape[2], X_train.shape[3]]),
    layers.MaxPool2D(),

    # Second Convolutional Block
    layers.Conv2D(filters=64, kernel_size=5, activation="relu", padding='same'),
    layers.MaxPool2D(),

    # Third Convolutional Block
    layers.Conv2D(filters=128, kernel_size=3, activation="relu", padding='same'),
    layers.MaxPool2D(),

    # Third Convolutional Block
    layers.Conv2D(filters=128, kernel_size=3, activation="relu", padding='same'),
    layers.MaxPool2D(),

    # Classifier Head
    layers.Flatten(),
    layers.Dense(units=512, activation="relu"),
    layers.Dense(units=256, activation="relu"),
    layers.Dense(units=1, activation="sigmoid"),
])
cnc_model.summary()
```

```python
cnc_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['binary_accuracy'],
)
```

```python
history = cnc_model.fit(
    X_train, Y_train,
    validation_data=(X_test, Y_test),
    batch_size=512,
    epochs=100,
    verbose=1,
)
```

```python
# График изменения точности
plt.plot(history.history['binary_accuracy'])
plt.plot(history.history['val_binary_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# График изменения функции потерь
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
Y_pred = cnc_model.predict(X_test) > 0.5 # Применили порог принятия решения
print(classification_report(Y_test, Y_pred))
print(confusion_matrix(Y_test, Y_pred))
```

- Визуализируем подобранные веса фильтров

```python
# Извлечем веса из второго скрытого слоя
filters, biases = cnc_model.layers[2].get_weights()

# Нормируем веса фильтров
f_min, f_max = filters.min(), filters.max()
filters = (filters - f_min) / (f_max - f_min)

# Визуализируем несколько фильтов
n_filters, ix = 6, 1
for i in range(n_filters):
	f = filters[:, :, :, i]
	# Для каждого канала
	for j in range(3):
		ax = plt.subplot(n_filters, 3, ix)
		ax.set_xticks([])
		ax.set_yticks([])
		plt.imshow(f[:, :, j], cmap='gray')
		ix += 1

plt.show()
```
